# 1. Test Random Environment with OpenAI Gym




```python
import gym
import random
from colabgymrender.recorder import Recorder
```

```python
#Game being used
env = gym.make('CartPole-v0')# -> gym.openai.com/#envs

#Google Colab rendering - comment if not needed;
env = Recorder(env, './video')

states = env.observation_space.shape[0]

#Number of actions available. In the current case there are only 2 (left or right) that will move the cart
actions = env.action_space.n
```

```python
print(states)
print(actions)
```

```python
#Env setting with random behaviour
episodes = 10
for episode in range(1,episodes+1):
    state = env.reset()
    done = False
    score = 0
    
    while not done:
        env.render(mode='rgb_array')
        action = random.choice([0,1])
        #print(env.step(action))
        n_state, reward, done, info = env.step(action)
        score+=reward
    print('Episode:{} Score:{}'.format(episode,score))

#Google Colab rendering - comment if not needed;
env.play()
```

# 2. Create a Deep Learning Model with Keras

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.optimizers import Adam
```

```python
def build_model(states, actions):
    model = Sequential()
    model.add(Flatten(input_shape=(1,states)))
    model.add(Dense(24,activation='relu'))
    model.add(Dense(24,activation='relu'))
    model.add(Dense(actions,activation='linear'))
    return model
```

```python
model = build_model(states,actions)
```

```python
model.summary()
```

```python
from rl.agents import DQNAgent #check diferent agents -> https://keras-rl.readthedocs.io/en/latest/
from rl.policy import BoltzmannQPolicy
from rl.memory import SequentialMemory
```

```python
def build_agent(model, actions):
    policy = BoltzmannQPolicy()
    memory = SequentialMemory(limit=50000,window_length=1)
    dqn = DQNAgent(model=model,memory=memory,policy=policy,
                  nb_actions=actions,nb_steps_warmup=10,target_model_update=1e-2)
    return dqn
```

```python
dqn = build_agent(model,actions)
dqn.compile(Adam(lr=1e-3), metrics=['mae'])
#I had to train the model with twice the number of steps(comparing to the tutorial) to achieve the expected result
dqn.fit(env,nb_steps=100000,visualize=False,verbose=1) 
```

```python
scores = dqn.test(env, nb_episodes=100, visualize=False)
print(np.mean(scores.history['episode_reward']))
```

# 4. Reload Agent from Memory

```python
dqn.save_weights('dqn_carpolev0_weights.h5f',overwrite=True)
```

```python
del model
del dqn 
del env
```

```python
'''
For testing we will try it out with Cartopole V1 now that terminates the episode 
after a longer time (500 points reached instead of 200), considering the model previous performance
we should be ok
'''
env = gym.make('CartPole-v1') 
#Google Colab rendering - comment if not needed;
env = Recorder(env, './video')

actions = env.action_space.n
states = env.observation_space.shape[0]
model = build_model(states, actions)
dqn = build_agent(model,actions)
dqn.compile(Adam(lr=1e-3),metrics=['mae'])
```

```python
dqn.load_weights('dqn_carpolev0_weights.h5f')
```
